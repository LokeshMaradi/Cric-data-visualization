```python
import pandas as pd
odi=pd.read_csv('ICC ODI Championship.csv')
odi
```

```python
import seaborn as sns
def trim(x):
    return x[:3]
```

```python
odi['Teams']=odi['Teams'].apply(trim)
```

```python
sns.barplot('Teams','Rating',data=odi)
```

```python
import matplotlib as mpl
mpl.rcParams['font.size']=25
```

```python
sns.barplot('Teams','Rating',data=odi)
```

```python
import matplotlib.pyplot as plt
p,ax=plt.subplots(figsize=(20,10))
ax=sns.barplot('Teams','Rating',data=odi)
```

```python
p,ax=plt.subplots(figsize=(20,10))
ax=sns.barplot('Rating','Teams',data=odi)
```

```python
wodi=pd.read_csv("ICC Women's ODI Team Rankings.csv")
```

```python
wodi['Teams']=wodi['Teams'].apply(trim)
```

```python
wodi['gender']='female'
odi['gender']='male'
```

```python
new=pd.concat([wodi,odi],ignore_index=True)
```

```python
p,ax=plt.subplots(figsize=(20,10))
ax=sns.barplot('Teams','Matches',hue='gender',data=new)
```
